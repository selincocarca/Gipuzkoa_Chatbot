# Descarga de los audios





Primero, instalamos la herramienta necesaria para descargar videos y audios de YouTube. Esto se hace ejecutando un comando que asegura que la biblioteca yt-dlp esté instalada y actualizada a la versión más reciente.

Luego, importamos las herramientas que vamos a usar en nuestro script. Necesitamos yt-dlp para descargar y convertir los audios, y os para manejar las carpetas y archivos en nuestro sistema.

Después, definimos una función que se encarga de descargar el audio de un video de YouTube. Esta función toma dos cosas: la URL del video y la carpeta donde queremos guardar el archivo. Dentro de la función, configuramos cómo queremos que se descargue el audio y en qué formato. En este caso, elegimos el formato MP3 con una calidad de 192 kbps.

A continuación, preparamos la carpeta donde guardaremos los audios. Si la carpeta ya existe, no hacemos nada; si no existe, la creamos.

Luego, especificamos una lista de videos de YouTube de los cuales queremos extraer el audio.

Finalmente, recorremos cada URL en nuestra lista y usamos la función definida para descargar y convertir el audio de cada video. Todo el audio descargado se guarda en la carpeta que preparamos al principio.

```python
# Instala la librería yt-dlp si aún no lo has hecho
!pip install --upgrade yt-dlp

# Importa las librerías necesarias
import yt_dlp
import os

# Función para descargar el audio de un video de YouTube
def descargar_audio(url, carpeta_destino):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': os.path.join(carpeta_destino, '%(title)s.%(ext)s'),
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'noplaylist': True,  # Evitar descargar listas de reproducción
        'force_generic_extractor': True,  # Forzar el uso del extractor genérico
        'extractor_args': {'youtube': {'nocheckcertificate': True}},
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

# Define la carpeta de destino para guardar los audios
carpeta_destino = "/content/audios"
os.makedirs(carpeta_destino, exist_ok=True)

# URLs de los videos de YouTube
urls = [
    "https://www.youtube.com/watch?v=SfB4jxtGyRo",
    "https://www.youtube.com/watch?v=H04knFkWPNE",
    "https://www.youtube.com/watch?v=AdPFyBC9lLw"
]

# Descarga y extrae el audio de cada video
for url in urls:
    descargar_audio(url, carpeta_destino)
```

# Transcripcion de los videos

Primero, instalamos las librerías necesarias para trabajar con audio y reconocimiento de voz si no las tienes ya instaladas. Usamos pydub para manejar archivos de audio, speechrecognition para transcribir el audio a texto, noisereduce para reducir el ruido en los audios, librosa para procesar audio, y soundfile para guardar archivos de audio.

Luego, importamos estas librerías en nuestro script. Cada una de ellas tiene un rol específico en el procesamiento del audio.

Ahora, definimos varias funciones para manejar el audio:

1. Reducir el Ruido: La primera función toma un archivo de audio, lo carga, le reduce el ruido y guarda una versión limpia en formato WAV.

2. Convertir a WAV: La segunda función convierte cualquier archivo de audio (como MP3) a formato WAV y lo guarda en una carpeta específica.

3. Dividir el Audio: La tercera función divide un archivo de audio en segmentos más pequeños de 30 segundos cada uno. Esto facilita el manejo del audio en trozos más manejables para la transcripción.

4. Transcribir el Audio: La última función toma un segmento de audio y usa un servicio de reconocimiento de voz para convertirlo en texto. Si el audio no es claro o hay un error, devuelve un mensaje indicando el problema.

A continuación, preparamos las carpetas donde guardaremos los archivos procesados. Creamos una carpeta para los archivos WAV si no existe ya.

Luego, comenzamos a procesar cada archivo de audio que está en la carpeta de audios. Para cada archivo MP3:

- Lo convertimos a formato WAV.
- Reducimos el ruido del archivo WAV.
- Dividimos el archivo de audio limpio en segmentos de 30 segundos.
- Transcribimos cada segmento a texto y juntamos todas las transcripciones en una sola cadena de texto.

Finalmente, mostramos la transcripción completa de cada archivo en la salida del notebook.

Este código automatiza todo el proceso de procesamiento de audio: desde la conversión y limpieza del audio hasta la transcripción y presentación de los resultados.

```python
# Instalar las librerías necesarias si no están instaladas
!pip install pydub speechrecognition noisereduce librosa soundfile

# Importar las librerías necesarias
import os
import librosa
import noisereduce as nr
import soundfile as sf
from pydub import AudioSegment
import speech_recognition as sr

# Función para reducir el ruido del audio
def reducir_ruido(ruta_audio):
    y, sr_rate = librosa.load(ruta_audio)
    y_reducido = nr.reduce_noise(y=y, sr=sr_rate)
    archivo_reducido = os.path.splitext(ruta_audio)[0] + "_reducido.wav"
    sf.write(archivo_reducido, y_reducido, sr_rate)
    return archivo_reducido

# Función para convertir audio a formato WAV
def convertir_a_wav(ruta_audio, carpeta_destino):
    audio = AudioSegment.from_file(ruta_audio)
    archivo_wav = os.path.join(carpeta_destino, os.path.splitext(os.path.basename(ruta_audio))[0] + '.wav')
    audio.export(archivo_wav, format="wav")
    return archivo_wav

# Función para dividir el audio en segmentos más pequeños
def dividir_audio(ruta_audio, duracion_segmento_ms=30000):
    audio = AudioSegment.from_wav(ruta_audio)
    segmentos = [audio[i:i+duracion_segmento_ms] for i in range(0, len(audio), duracion_segmento_ms)]
    return segmentos

# Función para transcribir audio a texto
def transcribir_audio_segmento(ruta_segmento):
    recognizer = sr.Recognizer()
    with sr.AudioFile(ruta_segmento) as source:
        audio = recognizer.record(source)
    try:
        texto = recognizer.recognize_google(audio, language="eu-ES")  # Cambiar el idioma si es necesario
        return texto
    except sr.UnknownValueError:
        return "[Inaudible]"
    except sr.RequestError as e:
        return f"[Error al solicitar resultados de reconocimiento de voz; {e}]"

# Definir las carpetas de destino
carpeta_audios = "/content/audios"
carpeta_wavs = "/content/wavs"
os.makedirs(carpeta_wavs, exist_ok=True)

# Procesar cada archivo de audio
for archivo in os.listdir(carpeta_audios):
    if archivo.endswith(".mp3"):
        ruta_audio = os.path.join(carpeta_audios, archivo)
        ruta_wav = convertir_a_wav(ruta_audio, carpeta_wavs)
        ruta_wav_reducido = reducir_ruido(ruta_wav)

        segmentos = dividir_audio(ruta_wav_reducido)
        transcripcion_completa = ""

        for i, segmento in enumerate(segmentos):
            # Guardar cada segmento como archivo WAV temporalmente
            archivo_segmento = f"/content/temp_segment_{i}.wav"
            segmento.export(archivo_segmento, format="wav")

            # Transcribir el segmento
            transcripcion = transcribir_audio_segmento(archivo_segmento)
            transcripcion_completa += transcripcion + " "

            # Eliminar el archivo de segmento temporal
            os.remove(archivo_segmento)

        # Mostrar la transcripción completa en la salida del notebook
        print(f"Transcripción del archivo {archivo}:\n")
        print(transcripcion_completa)
        print("\n" + "="*80 + "\n")
```

# Importación del PDF

Primero, instalamos la librería necesaria para manejar archivos PDF, en este caso, PyPDF2. Esta librería nos ayudará a extraer texto de archivos PDF.

Luego, importamos las librerías necesarias para trabajar con archivos y leer PDFs. Usamos os para manejar rutas de archivos y PyPDF2 para leer el contenido del PDF.

Para cargar un archivo desde tu ordenador a Google Colab, utilizamos una función de Google Colab llamada files.upload(). Esta función abre un diálogo que te permite seleccionar y subir archivos desde tu dispositivo.

Una vez que el archivo está subido, obtenemos una lista de los archivos subidos. En este caso, tomamos el primer archivo de la lista (asumiendo que solo has subido un archivo o que el archivo de interés es el primero en la lista).

Luego, definimos una función que convierte un archivo PDF a texto. Esta función abre el archivo PDF y utiliza PdfReader de PyPDF2 para leer cada página del PDF. Extrae el texto de cada página y lo concatena en una sola cadena de texto. Después, guarda el texto completo en un archivo de texto (.txt) en una carpeta específica.

Creamos una carpeta para almacenar los archivos de texto resultantes, asegurándonos de que la carpeta exista (si no existe, la creamos).

Finalmente, llamamos a la función para convertir el archivo PDF subido en texto y lo guardamos en la carpeta de transcripciones. Imprimimos la ruta donde se guardó el archivo de texto para que puedas acceder a él fácilmente.

En resumen, este código te permite subir un archivo PDF, convertir su contenido a texto y guardar ese texto en un archivo .txt en una ubicación específica dentro del entorno de Google Colab.

```python
# Instala las librerías necesarias
!pip install PyPDF2

# Importa las librerías necesarias
import os
from PyPDF2 import PdfReader

# Cargar un archivo desde tu ordenador a Google Colab
from google.colab import files
uploaded = files.upload()

# Obtiene la lista de archivos subidos y selecciona el primero (en caso de múltiples subidas)
uploaded_files = list(uploaded.keys())
ruta_pdf = uploaded_files[0]  # Tomamos el primer archivo subido

# Función para convertir PDF a texto
def convertir_pdf_a_texto(ruta_pdf, carpeta_destino):
    with open(ruta_pdf, "rb") as archivo:
        lector_pdf = PdfReader(archivo)
        texto_completo = ""
        for num_pagina in range(len(lector_pdf.pages)):
            pagina = lector_pdf.pages[num_pagina]
            texto_completo += pagina.extract_text()

    nombre_archivo_txt = os.path.splitext(os.path.basename(ruta_pdf))[0] + '.txt'
    ruta_txt = os.path.join(carpeta_destino, nombre_archivo_txt)
    with open(ruta_txt, "w", encoding="utf-8") as archivo_txt:
        archivo_txt.write(texto_completo)

    return ruta_txt

# Definir la carpeta de destino para guardar las transcripciones
carpeta_transcripciones = "/content/transcripciones"
os.makedirs(carpeta_transcripciones, exist_ok=True)

# Convertir el PDF a texto y guardar en la carpeta de transcripciones
ruta_txt = convertir_pdf_a_texto(ruta_pdf, carpeta_transcripciones)

# Mostrar la ruta del archivo de texto generado
print(f"El archivo de texto generado se encuentra en: {ruta_txt}")
```

   - La función `galderak_sortu` utiliza la API de OpenAI para crear preguntas basadas en el contenido del texto. Se generan un número especificado de preguntas a partir de los primeros 1000 caracteres del texto.

5. **Generar Respuestas**:
   - La función `erantzunak_sortu` usa la API de OpenAI para crear respuestas para cada una de las preguntas generadas, manteniendo el estilo y personalidad del euskera.

   - Se lee el contenido del archivo de texto.
   - Se generan preguntas a partir del texto leído.
   - Se generan respuestas para esas preguntas.
   - Finalmente, se imprimen las preguntas y respuestas generadas.

En resumen, el código lee un archivo de texto, genera preguntas basadas en ese texto usando la API de OpenAI, y luego genera y muestra respuestas a esas preguntas.

```python
import openai

# APIaren gakoaren konfigurazioa
openai.api_key = "tu api key"

def irakurri_testua_fitxategitik(fitxategi_bidea):
    """Irakurri testu-fitxategi baten edukia."""
    try:
        with open(fitxategi_bidea, 'r', encoding='utf-8') as fitxategia:
            return fitxategia.read()
    except Exception as e:
        print(f"Errore bat egon da fitxategia irakurtzean: {e}")
        return ""

def galderak_sortu(testua, galdera_kopurua=15):
    """Sortu galderak testuaren oinarrian OpenAI APIaren bidez."""
    sistemaren_edukia = """
    Testu baten oinarrian galderak sortzeko aditua zara. Galderak anitzak eta testuaren edukiarekin loturikoak izan behar dira. EZ ERREPIKATU GALDERAK.
    """

    testuaren_zati_1k = testua[:1000]  # Erabili testuaren lehen 1000 karaktereak

    erabiltzailearen_edukia = f"""
    Aginduak:
    1. Sortu {galdera_kopurua} galdera testu honen oinarrian:\n{testuaren_zati_1k}.
    2. Galderak testuaren kultura, gastronomia, historia eta beste alderdi garrantzitsuei buruzkoa izan behar dira.
    3. Galderak anitzak izan behar dira, eta ez da galdera bat ere errepikatu behar.
    4. Soil-soilik galderak inprimatu.
    """

    try:
        erantzuna = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": sistemaren_edukia},
                {"role": "user", "content": erabiltzailearen_edukia},
            ],
            temperature=0.7,
            max_tokens=500,
            top_p=0.7,
            frequency_penalty=0,
            presence_penalty=0
        )
        galderak = erantzuna.choices[0].message['content'].strip().split('\n')
        # Galderak hutsik daudenak eta errepikatuak ez direnak iragazi
        galderak = list(dict.fromkeys([g for g in galderak if g.strip()]))
        return galderak
    except Exception as e:
        print(f"Errore bat egon da galderak sortzean: {e}")
        return []

def erantzunak_sortu(testua, galderak):
    """Sortu erantzunak galderak oinarrituta OpenAI APIaren bidez, euskararen pertsonalitatea mantenduz."""
    galdera_erantzunak = []

    for galdera in galderak:
        try:
            erantzuna = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "Testu baten oinarrian galderak erantzuten espezializatuta zaude, euskararen pertsonalitatea eta adierazpenak mantenduz."},
                    {"role": "user", "content": f"Testua: {testua[:1000]}\n\nGaldera: {galdera}\n\nErantzun:"},
                ],
                temperature=1,
                max_tokens=300,
                top_p=1,
                frequency_penalty=0,
                presence_penalty=0
            )
            erantzuna_edu = erantzuna.choices[0].message['content'].strip()
            galdera_erantzunak.append((galdera, erantzuna_edu))
        except Exception as e:
            print(f"Errore bat egon da '{galdera}' galderaren erantzuna sortzean: {e}")

    return galdera_erantzunak

# Fitxategiaren bidea
fitxategi_bidea = 'transcripciones/GIPUZKOAKO GIDA.txt'

# Fitxategiaren edukia irakurri
testua = irakurri_testua_fitxategitik(fitxategi_bidea)

if testua:
    # Galderak sortu
    galdera_kopurua = 10
    galderak_sortuak = galderak_sortu(testua, galdera_kopurua)

    if galderak_sortuak:
        # Erantzunak sortu
        erantzunak_sortuak = erantzunak_sortu(testua, galderak_sortuak)

        # Emaitzak inprimatu
        for galdera, erantzuna in erantzunak_sortuak:
            print(f"Galdera (Euskera): {galdera}\nErantzun (Euskera): {erantzuna}\n")
    else:
        print("Ez dira galderak sortu.")
else:
    print("Testua ezin izan da irakurri.")
```

   - Si el trabajo de ajuste fino fue exitoso, se imprime el modelo ajustado (`status["fine_tuned_model"]`).

   - `evaluate_responses(expected_path, generated_path)`: Evalúa la precisión de las respuestas generadas en comparación con las respuestas esperadas.
     - **Leer Datos**:
       - Abre y lee el archivo de respuestas esperadas (`expected_path`), convirtiendo cada línea en un objeto JSON.
       - Abre y lee el archivo de respuestas generadas (`generated_path`), convirtiendo cada línea en un objeto JSON.
     
     - **Convertir Respuestas Generadas**:
       - Convierte las respuestas generadas en un diccionario `generated_dict` para un acceso rápido, usando `extract_message_content` para obtener la pregunta y la respuesta generada.

     - **Comparar Respuestas**:
       - Recorre las respuestas esperadas y compara cada pregunta con la respuesta generada correspondiente.
       - Cuenta cuántas respuestas generadas coinciden con las respuestas esperadas (`correct_count`).

     - **Calcular y Mostrar Precisión**:
       - Calcula la precisión como el porcentaje de respuestas correctas sobre el total de respuestas.
       - Imprime la precisión y las discrepancias entre respuestas esperadas y generadas, si las hay.

3. **Ejecutar Evaluación**:
   - Define las rutas de los archivos JSONL (`expected_path` y `generated_path`).
   - Llama a `evaluate_responses` para realizar la evaluación de precisión.

En resumen, el código compara respuestas generadas con respuestas esperadas desde archivos JSONL, calcula la precisión y muestra cualquier discrepancia encontrada.

```python
import json

def extract_message_content(messages):
    for message in messages:
        if message['role'] == 'user':
            question = message['content']
        elif message['role'] == 'assistant':
            generated_answer = message['content']
    return question, generated_answer

def evaluate_responses(expected_path, generated_path):
    # Leer las respuestas esperadas
    with open(expected_path, 'r') as file:
        expected_data = [json.loads(line) for line in file]

    # Leer las respuestas generadas
    with open(generated_path, 'r') as file:
        generated_data = [json.loads(line) for line in file]

    # Convertir las respuestas generadas en un diccionario para fácil acceso
    generated_dict = {}
    for item in generated_data:
        try:
            question, generated_answer = extract_message_content(item['messages'])
            generated_dict[question] = generated_answer
        except (KeyError, IndexError) as e:
            print(f"Error en una de las respuestas generadas: {e}")
            print(item)

    correct_count = 0
    total_count = len(expected_data)

    for item in expected_data:
        try:
            question = item['prompt']
            expected_answer = item['completion']
            generated_answer = generated_dict.get(question, '')

            if expected_answer.strip() == generated_answer.strip():
                correct_count += 1
            else:
                print(f"Pregunta: {question}")
                print(f"Respuesta esperada: {expected_answer}")
                print(f"Respuesta generada: {generated_answer}")
                print()
        except KeyError as e:
            print(f"Error en una de las respuestas esperadas: {e}")
            print(item)

    accuracy = (correct_count / total_count) * 100
    print(f"Precisión: {accuracy:.2f}%")

# Rutas a los archivos JSONL
expected_path = '/content/preguntas_respuestas.jsonl'
generated_path = '/content/preguntas_respuestas_chat.jsonl'

evaluate_responses(expected_path, generated_path)
```

# Indexación

Este código realiza la indexación de documentos en una colección de ChromaDB utilizando embeddings generados por el modelo `paraphrase-MiniLM-L6-v2` de `sentence-transformers`. A continuación se explica cada parte del código:

1. **Importaciones y Carga del Modelo de Embeddings**:
   - **`import chromadb`**: Importa la biblioteca ChromaDB.
   - **`from sentence_transformers import SentenceTransformer`**: Importa el modelo de embeddings de `sentence-transformers`.
   - **`model = SentenceTransformer('paraphrase-MiniLM-L6-v2')`**: Carga el modelo de embeddings que convierte texto en vectores numéricos.

2. **Inicialización del Cliente de ChromaDB**:
   - **`client = chromadb.Client()`**: Crea una instancia del cliente para interactuar con ChromaDB.

3. **Creación u Obtención de la Colección**:
   - **`collection_name = "my_collection"`**: Define el nombre de la colección.
   - **`try:` y `except chromadb.db.mixins.sysdb.UniqueConstraintError:`**: Intenta crear una nueva colección con el nombre dado. Si ya existe, la obtiene.
   - **`collection = client.create_collection(name=collection_name)`**: Crea la colección.
   - **`collection = client.get_collection(name=collection_name)`**: Obtiene la colección si ya existe.

4. **Funciones de Utilidad**:
   - **`read_text_file(filepath)`**: Lee el archivo de texto y retorna una lista de líneas.
   - **`get_vector_from_text(text)`**: Genera un vector a partir del texto utilizando el modelo de embeddings.
   - **`index_documents(filepath, batch_size=100)`**: Lee un archivo de texto, procesa sus líneas en lotes y las indexa en la colección de ChromaDB.

5. **Indexación de Documentos**:
   - **`lines = read_text_file(filepath)`**: Lee el archivo de texto.
   - **`for i, line in enumerate(lines):`**: Itera sobre las líneas del archivo.
   - **`doc_id = f"doc_{i}"`**: Asigna un identificador único a cada documento.
   - **`vector = get_vector_from_text(line.strip())`**: Genera un vector de embeddings para cada línea.
   - **`collection.add(ids=batch_ids, embeddings=batch_embeddings, metadatas=batch_metadatas)`**: Agrega los documentos a la colección en lotes.

6. **Ejecución de la Indexación**:
   - **`index_documents('/content/transcripciones/GIPUZKOAKO GIDA.txt')`**: Llama a la función de indexación con el archivo de texto especificado.

Este flujo permite almacenar y buscar documentos en ChromaDB utilizando representaciones vectoriales generadas a partir del contenido textual de los documentos.

```python
import chromadb
from sentence_transformers import SentenceTransformer

# Cargar el modelo de embeddings
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Crear una instancia del cliente de Chroma
client = chromadb.Client()

# Crear o obtener una colección
collection_name = "my_collection"
try:
    collection = client.create_collection(name=collection_name)
    print(f"Colección '{collection_name}' creada.")
except chromadb.db.mixins.sysdb.UniqueConstraintError:
    collection = client.get_collection(name=collection_name)
    print(f"Colección '{collection_name}' ya existe y ha sido obtenida.")

# Función para leer archivos de texto
def read_text_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        lines = file.readlines()
    return lines

# Función para obtener el vector del texto
def get_vector_from_text(text):
    return model.encode(text).tolist()

# Función para indexar documentos
def index_documents(filepath, batch_size=100):
    lines = read_text_file(filepath)
    batch_ids = []
    batch_embeddings = []
    batch_metadatas = []

    for i, line in enumerate(lines):
        doc_id = f"doc_{i}"
        vector = get_vector_from_text(line.strip())

        batch_ids.append(doc_id)
        batch_embeddings.append(vector)
        batch_metadatas.append({'text': line.strip()})

        # Si alcanzamos el tamaño del lote, añadimos los documentos a la colección
        if len(batch_ids) >= batch_size:
            try:
                collection.add(ids=batch_ids, embeddings=batch_embeddings, metadatas=batch_metadatas)
                print(f"Índice de documentos en lote {len(batch_ids)} completado.")
                batch_ids = []
                batch_embeddings = []
                batch_metadatas = []
            except Exception as e:
                print(f"Error al indexar el lote: {e}")

    # Añadir el último lote de documentos si hay alguno
    if batch_ids:
        try:
            collection.add(ids=batch_ids, embeddings=batch_embeddings, metadatas=batch_metadatas)
            print(f"Índice de documentos finalizado con los últimos {len(batch_ids)} documentos.")
        except Exception as e:
            print(f"Error al indexar el último lote: {e}")

# Ejecutar la indexación
index_documents('/content/transcripciones/GIPUZKOAKO GIDA.txt')
```

```python
import openai
import chromadb
from sentence_transformers import SentenceTransformer

# Configura tu API key de OpenAI
openai.api_key = 'tu api key'

# Cargar el modelo de embeddings
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Crear una instancia del cliente de Chroma
client = chromadb.Client()

# Nombre de la colección
collection_name = "my_collection"

# Crear o obtener una colección
def get_or_create_collection(name):
    try:
        collection = client.create_collection(name=name)
    except Exception as e:
        print(f"Error al crear colección: {e}. Intentando obtener colección existente.")
        try:
            collection = client.get_collection(name=name)
        except Exception as inner_e:
            print(f"Error al obtener colección existente: {inner_e}")
            collection = None
    return collection

collection = get_or_create_collection(collection_name)

# Función para obtener el vector del texto
def get_vector_from_text(text):
    try:
        return model.encode(text).tolist()
    except Exception as e:
        print(f"Error al obtener el vector del texto: {e}")
        return None

# Función para buscar documentos similares
def search_documents(query_text, top_k=5):
    query_vector = get_vector_from_text(query_text)
    if query_vector is None:
        print("No se pudo obtener el vector de la consulta.")
        return []

    try:
        results = collection.query(query_embeddings=[query_vector], n_results=top_k)
        if 'ids' not in results or not results['ids']:
            print("No se encontraron documentos relevantes.")
            return []

        retrieved_texts = []
        for ids in results['ids']:
            for doc_id in ids:
                metadata = collection.get(ids=[doc_id])['metadatas'][0]
                text = metadata['text']
                retrieved_texts.append(text)
        return retrieved_texts
    except Exception as e:
        print(f"Error en la búsqueda de documentos: {e}")
        return []

# Función para generar respuesta con GPT-3.5-turbo en euskera
def generate_response(query_text, retrieved_texts):
    if not retrieved_texts:
        return "Ez dago dokumenturik aurkitu kontsulta honetarako."

    # Crear un prompt más específico y claro
    prompt = f"""
    Kontsulta: "{query_text}"

    Aurkitu diren dokumentuak:
    {retrieved_texts}

    Dokumentu hauetan oinarrituta, kontsultari erantzun argi eta informativo bat sortu. Erantzunak honako hau izan behar du:
    - Dokumentuetan agertzen den informazioa laburbildu
    - Kontsultarekin lotutako gako-puntuak nabarmentzen
    """

    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "Zure lana dokumentuen oinarritutako erantzun argi eta informativoak sortzea da."},
                {"role": "user", "content": prompt}
            ]
        )
        return response.choices[0].message['content'].strip()
    except Exception as e:
        print(f"Error al generar la respuesta con OpenAI: {e}")
        return "Errore bat gertatu da erantzuna sortzerakoan."

# Ejecutar el pipeline
def run_pipeline(query_text):
    print("Iniciando la búsqueda...")
    retrieved_texts = search_documents(query_text)
    if not retrieved_texts:
        print("No se encontraron documentos relevantes.")
    print("Generando respuesta...")
    response = generate_response(query_text, retrieved_texts)
    print("Erantzun sortua:")
    print(response)

# Consultas de prueba ajustadas
test_queries = [
    "Zein da Pasaia herrian aurki daitekeen azken baleazalea?",
    "Nola deskribatzen da Gipuzkoaren paisai naturala?",
    "Zer da Gipuzkoaren itsas industria tradizionala?",
    "Nola deskribatzen dira Gipuzkoako mendiak?",
    "Zer da Gipuzkoaren historian misteriotsua?",
    "Nola deskribatzen da Gipuzkoaren itsasoa?",

]

# Ejecutar pruebas
for query in test_queries:
    print(f"Kontsulta probatu: {query}")
    run_pipeline(query)
    print("\n" + "-"*50 + "\n")
```

Esta aplicación de Streamlit proporciona una interfaz web para interactuar con un chatbot, utilizando tanto respuestas predefinidas como generadas por un modelo de lenguaje.

```python
import streamlit as st
import openai
import json
import os
from dotenv import load_dotenv

# Cargar las variables de entorno
load_dotenv()

# Configuración de la clave API de OpenAI desde la variable de entorno
openai.api_key = os.getenv("OPENAI_API_KEY")

# Archivos de almacenamiento
CHAT_HISTORY_FILE = 'chat_history.json'

def load_qa_pairs(filename):
    qa_pairs = {}
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            entry = json.loads(line.strip())
            question = entry['messages'][0]['content'].lower()
            answer = entry['messages'][1]['content']
            qa_pairs[question] = answer
    return qa_pairs

def get_gpt_answer(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": f"Erantzun euskaraz: {prompt}"}],
        max_tokens=150
    )
    return response.choices[0].message['content'].strip()

def get_answer(question):
    question = question.lower()
    return qa_pairs.get(question, get_gpt_answer(question))

def save_chat_history(chat_id, messages):
    chat_history = load_chat_history()
    chat_history[chat_id] = messages
    with open(CHAT_HISTORY_FILE, 'w', encoding='utf-8') as file:
        json.dump(chat_history, file, indent=4)
    print(f"Chat guardado: {chat_id} - Mensajes: {messages}")  # Depuración

def load_chat_history():
    if os.path.exists(CHAT_HISTORY_FILE):
        with open(CHAT_HISTORY_FILE, 'r', encoding='utf-8') as file:
            try:
                chat_history = json.load(file)
                print(f"Historial cargado: {chat_history}")  # Depuración
                return chat_history
            except json.JSONDecodeError:
                # Si el archivo está vacío o corrupto, inicializa un archivo nuevo
                print("Archivo corrupto o vacío, inicializando nuevo historial.")
                return {}
    else:
        # Inicializa el archivo si no existe
        with open(CHAT_HISTORY_FILE, 'w', encoding='utf-8') as file:
            json.dump({}, file)
        print("Archivo no existente, creado nuevo historial.")
        return {}

def delete_chat_history(chat_id):
    chat_history = load_chat_history()
    if chat_id in chat_history:
        del chat_history[chat_id]
        with open(CHAT_HISTORY_FILE, 'w', encoding='utf-8') as file:
            json.dump(chat_history, file, indent=4)
        st.success(f"Txat {chat_id} ezabatu da.")
    else:
        st.error(f"Txat {chat_id} ez dago historialean.")

# Inicializar pares de preguntas y respuestas
qa_pairs = load_qa_pairs('preguntas_respuestas_chat.jsonl')

# Configuración de la interfaz de usuario
st.set_page_config(page_title="Gipuzkoako Chatbot-a", page_icon=":robot_face:", layout="wide")

# Custom CSS
st.markdown("""
    <style>
    body {
      font-family: sans-serif;
      margin: 0;
      padding: 0;
      background-color: #f5f5f5;
    }
    .sidebar {
      background-color: #fff;
      width: 250px;
      padding: 20px;
      height: 100vh;
      overflow: auto;
      float: left;
    }
    .main {
      margin-left: 270px;
      padding: 20px;
      display: flex;
      flex-direction: column;
      height: 100vh;
      box-sizing: border-box;
    }
    .header {
      display: flex;
      justify-content: flex-start;
      align-items: flex-start;
      margin-bottom: 10px;
      padding: 10px;
      background-color: #e0e0e0;
      border-radius: 8px;
      box-sizing: border-box;
    }
    .logo {
      font-size: 24px;
      font-weight: bold;
      color: #333;
    }
    .chat-container {
      flex: 1;
      overflow-y: auto;
      display: flex;
      flex-direction: column-reverse;
      gap: 15px;
      margin-bottom: 10px;
    }
    .message {
      padding: 12px;
      border-radius: 10px;
      max-width: 60%;
      box-sizing: border-box;
      line-height: 1.5;
    }
    .user-message {
      background-color: #007bff;
      color: #fff;
      align-self: flex-end;
      text-align: right;
      border-top-right-radius: 0;
    }
    .assistant-message {
      background-color: #eee;
      color: #333;
      align-self: flex-start;
      border-top-left-radius: 0;
    }
    .input-container {
      margin-top: 10px;
      display: flex;
      flex-direction: column;
      gap: 10px;
    }
    textarea {
      width: 100%;
      height: 100px;
      padding: 10px;
      border: 1px solid #ddd;
      border-radius: 5px;
      font-size: 16px;
      box-sizing: border-box;
    }
    button {
      background-color: #007bff;
      color: #fff;
      padding: 10px 20px;
      border: none;
      border-radius: 5px;
      cursor: pointer;
      font-size: 16px;
      transition: background-color 0.3s, transform 0.3s, box-shadow 0.3s;
      box-shadow: 0 4px 8px rgba(0, 0, 0, 0.2);
    }
    button:hover {
      background-color: #0056b3;
      transform: scale(1.05);
    }
    button:focus {
      outline: none;
    }
    </style>
""", unsafe_allow_html=True)

# Función para iniciar un nuevo chat
def new_chat():
    # Guardar el historial actual solo si hay un chat activo
    if st.session_state.chat_id:
        save_chat_history(st.session_state.chat_id, st.session_state.messages)

    # Inicializar el nuevo chat
    st.session_state.messages = []
    st.session_state.user_input = ""
    st.session_state.chat_id = f"chat_{len(load_chat_history()) + 1}"

    # Inicializar el nuevo chat en el historial
    save_chat_history(st.session_state.chat_id, st.session_state.messages)
    # Eliminar la línea que muestra el mensaje de chat creado
    # st.write(f"Nuevo chat creado con ID: {st.session_state.chat_id}")

# Función para seleccionar un chat específico
def select_chat(chat_id):
    chat_history = load_chat_history()
    st.session_state.messages = chat_history.get(chat_id, [])
    st.session_state.chat_id = chat_id
    st.write(f"Chat seleccionado: {chat_id}")

# Manejo del estado del mensaje del usuario
if 'messages' not in st.session_state:
    st.session_state.messages = []
if 'user_input' not in st.session_state:
    st.session_state.user_input = ""
if 'chat_id' not in st.session_state:
    st.session_state.chat_id = None

# Sidebar
with st.sidebar:
    st.button("+ Txat Berria", on_click=new_chat)
    st.markdown("### Txat Historiala")
    chat_history = load_chat_history()
    chat_ids = list(chat_history.keys())

    selected_chat_id = st.selectbox("Aukeratu Txat bat:", ["-- Aukeratu --"] + chat_ids)
    if selected_chat_id != "-- Aukeratu --":
        if st.button(f"Txat {selected_chat_id} Ikusi", key=f"btn_view_{selected_chat_id}"):
            select_chat(selected_chat_id)
            st.experimental_rerun()

    if selected_chat_id != "-- Aukeratu --":
        if st.button(f"Txat {selected_chat_id} Ezabatu", key=f"btn_delete_{selected_chat_id}"):
            delete_chat_history(selected_chat_id)
            st.experimental_rerun()

# Mostrar el encabezado con el texto
st.markdown("""
    <div class='main'>
        <div class='header'>
            <div class='logo'>Gipuzkoako ChatBot-a</div>
        </div>
        <div class='chat-container'>
""", unsafe_allow_html=True)

# Mostrar el historial del chat actual
for message in st.session_state.messages:
    if message['role'] == 'user':
        st.markdown(f"<div class='message user-message'><strong>Zu:</strong> {message['content']}</div>", unsafe_allow_html=True)
    else:
        st.markdown(f"<div class='message assistant-message'><strong>Josune:</strong> {message['content']}</div>", unsafe_allow_html=True)

st.markdown("""
        </div>
        <div class='input-container'>
""", unsafe_allow_html=True)

# Campo de entrada de texto del usuario
user_input = st.text_area("Idatzi zure galdera hemen:", value=st.session_state.user_input, key="user_input_area")

# Botón para enviar el mensaje
if st.button("Bidali", key="send_button"):
    if user_input.strip():  # Asegúrate de que no sea solo espacio
        st.session_state.messages.append({"role": "user", "content": user_input})
        response = get_answer(user_input)
        st.session_state.messages.append({"role": "assistant", "content": response})
        st.session_state.user_input = ""

        # Guardar el historial del chat actual
        if st.session_state.chat_id:
            save_chat_history(st.session_state.chat_id, st.session_state.messages)

        st.experimental_rerun()

st.markdown("""
        </div>
    </div>
""", unsafe_allow_html=True)
```



1. **Importaciones y Configuración**:
   - Se importan las bibliotecas necesarias: `streamlit`, `openai`, `json`, y `pyngrok`.
   - Se configura la clave API de OpenAI.

2. **Funciones**:
   - **`load_qa_pairs(filename)`**: Carga pares de preguntas y respuestas desde un archivo JSONL.
   - **`get_gpt_answer(prompt)`**: Obtiene una respuesta de GPT-3.5-turbo para un prompt dado.
   - **`get_answer(question)`**: Devuelve la respuesta a una pregunta, buscando en los pares predefinidos y utilizando GPT si es necesario.

3. **Interfaz de Usuario**:
   - Configuración de la página en Streamlit (título, icono y diseño).
   - Manejo del estado de mensajes del usuario.
   - Muestra el historial de chat con estilos personalizados.
   - Campo de entrada para que el usuario escriba preguntas y un botón para enviar.

4. **Ngrok**:
   - **`start_ngrok()`**: Configura y devuelve una URL pública usando ngrok para exponer la aplicación.

5. **Guardar Código**:
   - El código se guarda en un archivo llamado `chatbot_app.py` en el directorio `/content/`.

Este script configura una aplicación web de chatbot con Streamlit, integra GPT-3.5-turbo, y utiliza ngrok para exponer la aplicación a Internet.

```python
# Crear el archivo chatbot_app.py
code = """
import streamlit as st
import openai
import json
from pyngrok import ngrok

# Configuración de la clave API de OpenAI
openai.api_key = "tu api key"

# Función para cargar pares de preguntas y respuestas desde un archivo JSONL
def load_qa_pairs(filename):
    qa_pairs = {}
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            entry = json.loads(line.strip())
            question = entry['messages'][0]['content'].lower()
            answer = entry['messages'][1]['content']
            qa_pairs[question] = answer
    return qa_pairs

# Inicializar pares de preguntas y respuestas
qa_pairs = load_qa_pairs('/content/preguntas_respuestas_chat.jsonl')

# Función para obtener la respuesta de GPT-3.5 Turbo
def get_gpt_answer(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=150
    )
    return response.choices[0].message['content'].strip()

# Función para responder preguntas
def get_answer(question):
    question = question.lower()
    return qa_pairs.get(question, get_gpt_answer(question))

# Configuración de la interfaz de usuario
st.set_page_config(page_title="Gipuzkoako Chatbot-a", page_icon=":robot_face:", layout="wide")




# Manejo del estado del mensaje del usuario
if 'messages' not in st.session_state:
    st.session_state.messages = []
if 'user_input' not in st.session_state:
    st.session_state.user_input = ""

# Mostrar el historial del chat
for message in st.session_state.messages:
    if message['role'] == 'user':
        st.markdown(f"<div class='message user-message'><strong>Zu:</strong> {message['content']}</div>", unsafe_allow_html=True)
    else:
        st.markdown(f"<div class='message assistant-message'><strong>Josune:</strong> {message['content']}</div>", unsafe_allow_html=True)

# Campo de entrada de texto del usuario
user_input = st.text_area("Idatzi zure galdera hemen:", value=st.session_state.user_input, key="user_input_area")

# Botón para enviar el mensaje
if st.button("Bidali", key="send_button"):
    if user_input.strip():  # Asegúrate de que no sea solo espacio
        # Guardar el mensaje del usuario
        st.session_state.messages.append({"role": "user", "content": user_input})

        # Obtener respuesta del chatbot
        response = get_answer(user_input)

        # Guardar la respuesta del asistente
        st.session_state.messages.append({"role": "assistant", "content": response})

        # Limpiar el campo de entrada
        st.session_state.user_input = ""  # Limpiar el estado del campo de entrada
        st.experimental_rerun()  # Actualizar la página



# Configurar el túnel ngrok
def start_ngrok():
    public_url = ngrok.connect(8501)
    return public_url

# Iniciar ngrok y mostrar URL pública
public_url = start_ngrok()
st.write(f"Tu URL pública es: {public_url}")

"""

with open('/content/chatbot_app.py', 'w') as file:
    file.write(code)
```
